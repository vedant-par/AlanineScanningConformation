```python
sequence = "NVTNLCPFHEVFNATRFASVYAWNRTRISNCVADYSVLYNFAPFFAFKCYGVSPTKLNDLCFTNVYADSFVIKGNEVSQIAPGQTGNIADYNYKLPDDFTGCVIAWNSNKLDSKHSGNYDYWYRSRKSKLKPFERDISTEIYQAGNKPCKGKGPNCYFPLESYGFRPTYGVGHQPYRVVVLSFELLHAPATVCGPKKS"
```

```python
import random
```

```python
seqs = []
```

```python
newseq = ""
```

```python
for j in range(100):
  temp = random.uniform(0.05, 0.15)
  for i in range(len(sequence)):
    if random.random() < temp:
      newseq += "A"
    else:
      newseq += sequence[i]
  seqs.append(newseq)
  newseq = ""
```
